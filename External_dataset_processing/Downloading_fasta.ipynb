# Downloading genome fasta files from other studies
I need to download the fasta files for the genomes from the other studies then I will annotate them using prokka

```python
import os

workDir = '/home/sam/FullCyc_metagenome/Other_studies_comp/redo_analysis'
```

```python
import pandas as pd
import re
from Bio import SeqIO
```

## Download fasta files

### Rainfall dataset

```python
acc_list = [re.sub('\n$', '', line) for line in open(os.path.join(workDir, 'accession_lists', 'rainfall_acc.txt'), 'r').readlines()]
acc_list
```

```python
for acc in acc_list:
    acc_let = acc.split('_')[0]
    acc_num = acc.split('_')[1].split('.')[0]
    acc_v = acc.split('_')[1].split('.')[1]
    acc_asm = re.sub('.$', '', re.sub('^0*', 'ASM', acc.split('_')[1].split('.')[0])) + 'v' + acc_v
    html_path = os.path.join('https://ftp.ncbi.nlm.nih.gov/genomes/all', 
                             acc_let, 
                             acc_num[0:3],
                             acc_num[3:6],
                             acc_num[6:9],
                             acc + '_' + acc_asm,
                             acc + '_' + acc_asm + '_genomic.fna.gz')
    outpath = os.path.join(workDir, 'genome_fasta', 'rainfall')
    cmd = ' '.join(['wget', 
                    '--directory-prefix', outpath,
                    html_path])
    !$cmd
```

```python
print('There should be: ' + str(len(acc_list)))
print('There are: ' + str(len(os.listdir(os.path.join(workDir, 'genome_fasta', 'rainfall')))))
```

## RefSoil

```python
import os
import pandas as pd
from Bio import Entrez
from urllib.request import HTTPError
import time
```

```python
RefSoil_metadata = pd.read_csv('/home/sam/new_databases/RefSoil/RefSoil_genomes/RefSoil_metadata.txt', sep='\t')
RefSoil_metadata
```

```python
RefSoil_acc = list(RefSoil_metadata['NCBI.ID'])
```

```python
Entrez.email = 'seb369@cornell.edu'
for acc in RefSoil_acc:
    try:
        handle = Entrez.efetch(db="nucleotide", id=acc, rettype='fasta', retmode='text')
    except HTTPError:
        time.sleep(20)
        handle = Entrez.efetch(db="nucleotide", id=acc, rettype='fasta', retmode='text')
    with open(os.path.join('/home/sam/new_databases/RefSoil/RefSoil_genomes/genome_fasta', acc + '.fna'), 'w') as outfile:
        outfile.write(handle.read())
    time.sleep(1)
```

## DeepSIP
We need to rename the contigs for this study

```python
orig_dir = '/home/sam/FullCyc_metagenome/Other_studies_comp/DeepSIP/mag_DNA_fasta'
new_dir = os.path.join(workDir, 'genome_fasta', 'DeepSIP')

for fa in [fa for fa in os.listdir(orig_dir) if fa.endswith('.fa')]:
    scaffold = 1
    with open(os.path.join(new_dir, fa), 'w') as outfile:
        for record in SeqIO.parse(os.path.join(orig_dir, fa), 'fasta'):
            outfile.write('>scaffold_' + str(scaffold) + '\n')
            outfile.write(str(record.seq) + '\n')
            scaffold += 1
    
```
