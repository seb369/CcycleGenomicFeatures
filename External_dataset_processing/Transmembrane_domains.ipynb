# Running TMHMM

```python
import os
workdir = '/home/sam/FullCyc_metagenome/Other_studies_comp/redo_analysis/TMHMM_output'
```

```python
import pandas as pd
import multiprocessing
import re
import time
```

```python
def run_tmhmm(infaa):
    outfile = os.path.join(outdir, os.path.split(infaa)[1]) + '.tmhmm'
    cmd = ' '.join(['tmhmm',
                    infaa, 
                    '-short',
                    '-noplot',
                    '>', outfile])
    os.system(cmd)
```

# RefSoil dataset

```python
prokka_dir = '/home/sam/FullCyc_metagenome/Other_studies_comp/redo_analysis/prokka_output/RefSoil'
faa_list = [os.path.join(prokka_dir, genome, genome + '.prokka.faa') for genome in os.listdir(prokka_dir)]

outdir = os.path.join(workdir, 'RefSoil')

if len(faa_list) < 70:
    nproc = len(faa_list)
else:
    nproc = 70
        
print(len(faa_list))
```

```python
pool = multiprocessing.Pool(processes=nproc)
pool.map(run_tmhmm, faa_list)
```

```python
#print('Ran for ', str(round(run_time/60, 3)) + ' minutes')
print('There should be: ' + str(len(faa_list)))
print('There are: ' + str(len(os.listdir(outdir))))
```
