# Annotating genomes from other studies using PROKKA
I'll use PROKKA to annotate the MAGs from the other studies

```python
import os

workDir = '/home/sam/FullCyc_metagenome/Other_studies_comp/redo_analysis'
```

```python
import pandas as pd
import re
import multiprocessing
import time
```

```python
def run_prokka(inputfile):
    cmd = ' '.join(['prokka', '--quiet',
                        '--cpus 1',
                        '--outdir', os.path.join(outdir, inputfile),
                        '--prefix', inputfile + '.prokka',
                        os.path.join(indir, inputfile)])
    os.system(cmd)
```

## Rainfall dataset

```python
fna_dir = os.path.join(workDir, 'genome_fasta', 'rainfall')
fna_list = [fna for fna in os.listdir(fna_dir)]

indir = os.path.join(workDir, 'genome_fasta', 'rainfall')
outdir = os.path.join(workDir, 'prokka_output', 'rainfall')

if len(fna_list) < 70:
    nproc = len(fna_list)
else:
    nproc = 70

len(fna_list)
```

```python
fna_list
```

```python
pool = multiprocessing.Pool(processes=nproc)
pool.map(run_prokka, fna_list)
```

```python
print('There should be: ' + str(len(fna_list)))
print('There are: ' + str(len(os.listdir(os.path.join(workDir, 'prokka_output', 'rainfall')))))
```
