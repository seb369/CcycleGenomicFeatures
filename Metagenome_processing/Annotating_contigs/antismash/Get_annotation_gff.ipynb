# Getting gff3 files of Prodigal annotations

To run ANTIsmash, I need to get Prodigal annotations for just the enriched contigs. The Antismash input can take GFF3 files so I need to subset the file generated by IMG.

```python
import os
import pandas as pd

workDir = '/home/sam/FullCyc_metagenome/annotation/antismash/'
enr_contig_df = '/home/sam/FullCyc_metagenome/enriched_binning_10_20/enriched_contig_groups.txt'
contig_fasta = '/home/sam/FullCyc_metagenome/contigs/assembly.contigs.1000.fasta'
prod_gff = '/home/sam/FullCyc_metagenome/annotation/IMG/Ga0334612_prodigal.gff'
IMG_gff = '/home/sam/FullCyc_metagenome/annotation/IMG/Ga0334612_functional_annotation.gff'

ncores = 15
```

## Make gff

```python
enr_contigs = list(set(pd.read_csv(enr_contig_df, sep='\t')['contigName']))
len(enr_contigs)
```

```python
contigs2scaffolds = pd.read_csv('/home/sam/FullCyc_metagenome/annotation/IMG/Ga0334612_contig_names_mapping.tsv', sep='\t', header=None)
contigs2scaffolds.columns = ['contigName', 'Scaffold']
enr_scaffolds = list(set(contigs2scaffolds[contigs2scaffolds.contigName.isin(enr_contigs)]['Scaffold']))
len(enr_scaffolds)
```

```python
with open(os.path.join(workDir, 'enr_scaffold_list.txt'), 'w') as enr_scaffold_file:
    enr_scaffold_file.write('\n'.join(enr_scaffolds))
```

```python
cmd = 'head -n 1 ' + prod_gff + ' > ' + os.path.join(workDir, 'enr_contigs_prodigal.gff')
os.system(cmd)
cmd = 'grep -Ff ' + os.path.join(workDir, 'enr_scaffold_list.txt') + ' ' + IMG_gff + ' >> ' + os.path.join(workDir, 'enr_contigs_prodigal.gff')
os.system(cmd)


```
